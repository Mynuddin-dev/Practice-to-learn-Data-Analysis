```python
# Code as a hobby
with open("survey_results_public.csv" , encoding="utf-8") as f:
    csv_reader = csv.DictReader(f)
    
    count_yes = 0
    count_no = 0
    count_na = 0
    
    for line in csv_reader:
        
        if line['Hobbyist'] == 'Yes':
            count_yes += 1
        elif line['Hobbyist'] == 'No':
            count_no += 1
        elif line['Hobbyist'] == 'NA':
            count_na += 1
            
total = count_yes + count_no + count_na
count_no_per = (count_no/total)*100
count_yes_per = (count_yes/total)*100
count_na_per = (count_na/total)*100

    
print(f"Yes : {round(count_yes_per , 2)}%")
print(f"No : {round(count_no_per , 2)}%")
print(f"NA : {round(count_na_per , 2)}%")
```

```python
# Most Popular language

with open("survey_results_public.csv" , encoding="utf-8") as f:
    csv_reader = csv.DictReader(f)
    
    language_counts = Counter()
    total = 0
    for line in csv_reader:
        
        languagesss = line["LanguageWorkedWith"].split(";")
        
        #language_counts.update(languagesss)
        for language in languagesss:
            language_counts[language] += 1
            
        total += 1
        
for lang , value in language_counts.most_common(5):
    
    count_language_per = (value/total)*100
    print(f"{lang} : {round(count_language_per , 2)}%")
    
print(total)    
    
    
```

```python
# most popular language based on dveloper type
import csv
from collections import defaultdict, Counter

with open("survey_results_public.csv" , encoding="utf-8") as f:
    csv_reader = csv.DictReader(f)

    dev_type_info = {}

    for line in csv_reader:
        dev_types = line['DevType'].split(';')

        for dev_type in dev_types:
            dev_type_info.setdefault(dev_type, {
                'total': 0,
                'language_counter': Counter()
            })

            languages = line['LanguageWorkedWith'].split(';')
            dev_type_info[dev_type]['language_counter'].update(languages)
            dev_type_info[dev_type]['total'] += 1


for dev_type, info in dev_type_info.items():
    print(dev_type)

    for language, value in info['language_counter'].most_common(5):
        language_pct = (value / info['total']) * 100
        language_pct = round(language_pct, 2)

        print(f'\t{language}: {language_pct}%')
```
