 

<p> <b>A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.</b>

Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables. They are particularly adept at showing interactions: how the relationship between levels of one categorical variable changes across levels of a second categorical variable. The lines that join each point from the same hue level allow interactions to be judged by differences in slope, which is easier for the eyes than comparing the heights of several groups of points or bars.

<b>It is important to keep in mind that a point plot shows only the mean (or other estimator) value</b>, but in many cases it may be more informative to show the distribution of values at each level of the categorical variables. In that case, other approaches such as a box or violin plot may be more appropriate
 
<b>Use catplot() to combine a pointplot() and a FacetGrid.</b> This allows grouping within additional categorical variables. Using catplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets:

</p>

<h3> Using pointplot </h3>

```python
# Draw a set of vertical point plots grouped by a categorical variable:

sns.pointplot(x="time", y="total_bill", data=tips)
```

```python

sns.set(style="darkgrid")
sns.pointplot(x="time", y="total_bill", data=tips)
```

```python
# Draw a set of vertical points with nested grouping by a two variables:

sns.pointplot(x="time", y="total_bill", hue="smoker", data=tips)
```

```python
# Separate the points for different hue levels along the categorical axis:

sns.pointplot(x="time", y="total_bill", hue="smoker", data=tips, dodge=True)
```

```python
#Use a different marker and line style for the hue levels:

sns.pointplot(x="time", y="total_bill", hue="smoker", data=tips,markers=["o", "x"],linestyles=["-", "--"])
```

```python
#Draw a set of horizontal points:

sns.pointplot(x="tip", y="day", data=tips)
```

```python
# Don’t draw a line connecting each point:

sns.pointplot(x="tip", y="day", data=tips, join=False)
```

```python
# Use a different color for a single-layer plot:

ax = sns.pointplot("time", y="total_bill", data=tips,
                   color="#bb3f3f")
```

```python
# Use a different color palette for the points:

ax = sns.pointplot(x="time", y="total_bill", hue="smoker", data=tips, palette="Set2")
```

```python
# Control point order by passing an explicit order:

sns.pointplot(x="time", y="tip", data=tips , order=["Dinner", "Lunch"])
```

```python
# Use median as the estimate of central tendency:

from numpy import median
sns.pointplot(x="day", y="tip", data=tips, estimator=median)
```

```python
# Show the standard error of the mean with the error bars:

sns.pointplot(x="day", y="tip", data=tips, ci=68)
```

```python
#Show standard deviation of observations instead of a confidence interval:

sns.pointplot(x="day", y="tip", data=tips, ci="sd")
```

```python
# Show the standard error of the mean with the error bars:

ax = sns.pointplot(x="day", y="tip", data=tips, ci=68)
```

```python
# Add “caps” to the error bars:

sns.pointplot(x="day", y="tip", data=tips, capsize=.2)
```
