
<p>A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

This can be an effective and attractive way to show multiple distributions of data at once, but keep in mind that the estimation procedure is influenced by the sample size, and violins for relatively small samples might look misleadingly smooth.<br>

Use catplot() to combine a violinplot() and a FacetGrid. This allows grouping within additional categorical variables. Using catplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets:</p>

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

tips = pd.read_csv("tips.csv")
tips.head()
```

```python
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="violin", split=True,height=4) #aspect=.7
```

```python
sales = pd.read_excel("Sample - Superstore Sales.xls")
sales.head()
```

```python
sns.catplot(x = "Product Category" , y = "Shipping Cost" , data = sales ,kind = "violin")
```

```python
sns.catplot(x = "Product Category" , y = "Shipping Cost" , hue = "Ship Mode" , data = sales ,kind = "violin")
```

```python
sns.catplot(x = "Product Category" , y = "Shipping Cost" , hue = "Ship Mode" , data = sales ,kind = "box")
sns.catplot(x = "Product Category" , y = "Shipping Cost" , hue = "Ship Mode" , data = sales ,kind = "violin")
```

```python
sns.catplot(x = "Product Category" , y = "Shipping Cost" , hue = "Ship Mode" , inner = "stick" , data = sales ,kind = "violin")
```


<h3> Use Violin </h3>

```python
sns.violinplot(x=tips["total_bill"])
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips)
```

```python
# Draw a violinplot with nested grouping by two categorical variables:

sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips, palette="muted")
```

```python
# Draw split violins to compare the across the hue variable:

sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips, split=True)
```

```python
# Control violin order by passing an explicit order:

sns.violinplot(x="time", y="tip", data=tips,
                    order=["Dinner", "Lunch"])
```

```python
# Scale the violin width by the number of observations in each bin:

ax = sns.violinplot(x="day", y="total_bill", hue="sex" , data=tips, palette="Set1", split=True, scale="count")
```

```python
#Scale the violin width by the number of observations in each bin:
sns.violinplot(x="day", y="total_bill", hue="sex",data=tips, palette="Set1", split=True)
```

```python
# Draw the quartiles as horizontal lines instead of a mini-box:

sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set1", split=True,scale="count", inner="quartile")


sns.catplot(x="day", y="total_bill", hue="sex",
                    data=tips, palette="Set1", split=True,
                    scale="count", inner="quartile" , kind  = "violin")
```

```python
#Show each observation with a stick inside the violin:

#sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set1", split=True,scale="count", inner="stick")

sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set1", split=True,scale="count", inner="quartile")

sns.catplot(x="day", y="total_bill", hue="sex",data=tips, palette="Set1", split=True,scale="count", inner="stick" , kind  = "violin")
```

```python
# Use a narrow bandwidth to reduce the amount of smoothing:
# Scale the density relative to the counts across all bins:

sns.violinplot(x="day", y="total_bill", hue="sex",
                    data=tips, palette="Set2", split=True,
                    scale="count", inner="stick",
                    scale_hue=False, bw=.2)
```

```python
sns.violinplot(x="total_bill", y="day", hue="sex",
                    data=tips, palette="Set2", split=True,
                    scale="count", inner="stick",
                    scale_hue=False, bw=.2)
```
