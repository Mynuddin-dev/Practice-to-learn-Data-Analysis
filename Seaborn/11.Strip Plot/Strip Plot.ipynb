
<p>Draw a scatterplot where one variable is categorical.

A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.<br><b>Everything is under catplot </b></p>

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

tips = pd.read_csv("tips.csv")
tips.head()
```

```python
tips.columns
```

```python
sns.stripplot(x= tips["tip"])
```

```python
sns.stripplot(x= tips["total_bill"])
```

```python
sns.stripplot(x= tips["tip"] , color = "green")
```

```python
#Group the strips by a categorical variable:

sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
#Use a smaller amount of jitter:

sns.stripplot(x="day", y="total_bill", data=tips, jitter=0.05)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=0.2)
```

```python
# Draw horizontal strips:

sns.stripplot(x="total_bill", y="day", data=tips)
```

```python
sns.stripplot(x="total_bill", y="day", data=tips , jitter = 1)
```

```python
# Draw outlines around the points:

sns.stripplot(x="total_bill", y="day", data=tips , linewidth=1)
```

```python
# Nest the strips within a second categorical variable:

sns.stripplot(x="sex", y="total_bill", hue="day", data=tips)
```

```python
sns.catplot(x="sex", y="total_bill", hue="day", data=tips , kind = "strip")
```

```python
# Draw each level of the hue variable at different locations on the major categorical axis:

sns.stripplot(x="day" , y="total_bill" , hue="smoker" , data=tips , palette="Set2" , dodge=True)
```

```python
sns.stripplot(x="day" , y="total_bill" , hue="smoker" , data=tips , palette="Set2")
```

```python
# Control strip order by passing an explicit order:

sns.stripplot(x="time", y="tip", data=tips , order=["Dinner", "Lunch"])
```

```python
#Draw strips with large points and different aesthetics:

sns.stripplot("day", "total_bill", "smoker", data=tips, palette="Set2", size=20, marker="D",edgecolor="gray", alpha=.5)
```

```python
# Draw strips of observations on top of a box plot:

import numpy as np
sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)      # not clear about this part
sns.stripplot(x="tip", y="day", data=tips, color=".3")
```

```python
import numpy as np
sns.boxplot(x="tip", y="day", data=tips)     
sns.stripplot(x="tip", y="day", data=tips, color=".3")
```

```python
# Draw strips of observations on top of a violin plot:

ax = sns.violinplot(x="day", y="total_bill", data=tips,
                    inner=None, color=".8")
ax = sns.stripplot(x="day", y="total_bill", data=tips)
```
