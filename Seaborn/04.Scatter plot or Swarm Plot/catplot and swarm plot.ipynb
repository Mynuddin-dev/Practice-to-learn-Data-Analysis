```python
# In scatter plot using numerical value to creat scatter chart

# Draw a categorical scatterplot with non-overlapping points.

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

sales = pd.read_excel("Sample - Superstore Sales.xls")
sales.head()
```

```python
sns.catplot("Ship Mode" , "Sales" , data = sales)
```

```python
sns.catplot("Ship Mode" , "Sales" , data = sales , jitter = False)
```

```python
sns.catplot("Ship Mode" , "Sales" , data = sales.query("Profit > 1000") , jitter = False)
```

```python
sns.catplot("Ship Mode" , "Sales" , data = sales.query("Profit > 1000") , kind = "swarm")
```

```python
sns.catplot("Sales" ,"Ship Mode" , data = sales.query("Profit > 1000") , kind = "swarm")
```

```python
tips = pd.read_csv("tips.csv")
tips
```

```python
# Use catplot() to combine a swarmplot() and a FacetGrid. This allows grouping within additional categorical variables.
# Using catplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets:



# Use catplot() is safe and just use kind = swarm

g = sns.catplot(x="sex", y="total_bill",hue="smoker", col="time", data=tips, kind="swarm", height=4, aspect=.7);

# If you dont want to use this way there you can use swarmplot

#sns.swarmplot(x=tips["total_bill"])
```

```python
#Group the swarms by a categorical variable:

sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
#Draw horizontal swarms:

ax = sns.swarmplot(x="total_bill", y="day", data=tips)
```

```python
# Color the points using a second categorical variable:

sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", data=tips , kind = "swarm")   # thats here is differance ts clear
```

```python
# Split each level of the hue variable along the categorical axis:

ax = sns.swarmplot(x="day", y="total_bill", hue="smoker",
                   data=tips, palette="Set2", dodge=True)
```

```python

sns.catplot(x="day", y="total_bill", hue="smoker",data=tips, palette="Set2", dodge=True , kind = "swarm")
```

```python
# Control swarm order by passing an explicit order:

sns.swarmplot(x="time" , y="tip" , data=tips , order=["Dinner", "Lunch"])
```

```python
# Plot using larger points:

ax = sns.swarmplot(x="time", y="tip", data=tips, size=6)
```

```python
#Draw swarms of observations on top of a box plot:

ax = sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)
ax = sns.swarmplot(x="tip", y="day", data=tips, color=".2")
```

```python
# Draw swarms of observations on top of a violin plot:

ax = sns.violinplot(x="day", y="total_bill", data=tips, inner=None )
ax = sns.swarmplot(x="day", y="total_bill", data=tips,
                   color="white", edgecolor="gray" )
```
