
<h3><a href = "https://seaborn.pydata.org/generated/seaborn.jointplot.html"> Seaborn Jointplot </a> </h3>
<p>Draw a plot of two variables with bivariate and univariate graphs.

This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly.</p>

```python
sns.jointplot(x = "Sales" , y= "Product Base Margin" , data = sales)
```

```python
sns.jointplot(x = "Sales" , y= "Product Base Margin" , data = sales , kind =  "hex")
```

```python
sns.jointplot(x = "Discount" , y= "Product Base Margin" , data = sales , kind =  "kde")
```

```python
# From jointplot.org

#Draw a scatterplot with marginal histograms

sns.set(style="white", color_codes=True)

tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
#Add regression and kernel density fits:

g = sns.jointplot("total_bill", "tip", data=tips, kind="reg")
```

```python
#Replace the scatterplot with a joint histogram using hexagonal bins:


sns.jointplot("total_bill", "tip", data=tips, kind="hex")
```

```python
# Replace the scatterplots and histograms with density estimates and align the marginal Axes tightly with the joint Axes:


sns.jointplot("total_bill", "tip", data=tips, kind="kde", space=0, color="g")
```

```python
# Draw a scatterplot, then add a joint density estimate:

sns.jointplot("total_bill", "tip", data = tips, color="k").plot_joint(sns.kdeplot, zorder=0, n_levels=6)
```

```python
# Draw a smaller figure with more space devoted to the marginal plots:

sns.jointplot("total_bill", "tip", data=tips , height=5, ratio=3, color="g")
```

```python
#Pass keyword arguments down to the underlying plots:

sns.jointplot("total_bill", "tip", data=tips,marginal_kws=dict(bins=15, rug=True),
                                  annot_kws=dict(stat="r"), s=40, edgecolor="w", linewidth=1)
```
