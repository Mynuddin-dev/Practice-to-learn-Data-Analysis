
<p> <b>A bar plot represents an estimate of central tendency for a numeric variable</b> with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars. Bar plots include 0 in the quantitative axis range, and they are a good choice when 0 is a meaningful value for the quantitative variable, and you want to make comparisons against it.

For datasets where 0 is not a meaningful value, a point plot will allow you to focus on differences between levels of one or more categorical variables.

It is also important to keep in mind that<b> a bar plot shows only the mean (or other estimator) value</b>, but in many cases it may be more informative to show the distribution of values at each level of the categorical variables. In that case, other approaches such as a box or violin plot may be more appropriate.
    
<b>Use catplot() to combine a barplot() and a FacetGrid.</b> This allows grouping within additional categorical variables. Using catplot() is safer than using FacetGrid directly, as it ensures synchronization of variable order across facets:

</p>

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

tips = pd.read_csv("tips.csv")
tips.head()
```

```python
 sns.catplot(x="sex", y="total_bill",hue="smoker", col="time",data=tips, kind="bar")
```

```python
 sns.catplot(x="sex", y="total_bill",hue="smoker", col="time",data=tips, kind="bar",height=4)
```

```python
 sns.catplot(x="sex", y="total_bill",hue="smoker", col="time",data=tips, kind="bar",height=4, aspect=.7)
```

```python
 sns.catplot(x="sex", y="total_bill",hue="smoker", col="time",data=tips, kind="bar",aspect=.7)
```

<h2> Use Barplot </h2>

```python
# Draw a set of vertical bar plots grouped by a categorical variable:

import seaborn as sns
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")                         # this is the another way of load dataset
sns.barplot(x="day", y="total_bill", data=tips)
tips.head()
```

```python
#Draw a set of vertical bars with nested grouping by a two variables:

sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
sns.catplot(x="day", y="total_bill", hue="sex", data=tips , kind = "bar")
```

```python
# Draw a set of horizontal bars:

sns.barplot(x="tip", y="day", data=tips)
```

```python
#Control bar order by passing an explicit order:

sns.barplot(x="time", y="tip", data=tips,
                 order=["Dinner", "Lunch"])
```

```python
#Use median as the estimate of central tendency:

from numpy import median
sns.barplot(x="day", y="tip", data=tips, estimator=median)
```

```python
# Show the standard error of the mean with the error bars:

sns.barplot(x="day", y="tip", data=tips, ci=68)   #Size of confidence intervals to draw around estimated values
```

```python
# Show standard deviation of observations instead of a confidence interval:

sns.barplot(x="day", y="tip", data=tips, ci="sd")
```

```python
# Add “caps” to the error bars:

ax = sns.barplot(x="day", y="tip", data=tips, capsize=.2)
```

```python
# Use a different color palette for the bars:

ax = sns.barplot("size" , y="total_bill" , data=tips , palette="Blues_d")
```

```python
ax = sns.barplot("size" , y="total_bill" , data=tips , palette="Reds_d")
```

```python
#Plot all bars in a single color:

sns.barplot("size", y="total_bill", data=tips,color="salmon", saturation=.5)
```
