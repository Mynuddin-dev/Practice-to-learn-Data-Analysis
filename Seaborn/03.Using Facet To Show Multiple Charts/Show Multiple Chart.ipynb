```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

sales = pd.read_excel("Sample - Superstore Sales.xls")
sales.head()
```

```python
sns.relplot(x= "Sales" , y= "Profit" , data = sales)
```

```python
sns.relplot(x= "Sales" , y= "Profit" , data = sales , col ="Customer Segment")
```

```python
sns.relplot(x= "Sales" , y= "Profit" , data = sales , col ="Customer Segment" , col_wrap = 4)
```

```python
sns.relplot(x= "Sales" , y= "Profit" , data = sales , col ="Customer Segment" , col_wrap = 2)
```

```python
sns.relplot(x= "Sales" , y= "Profit" , data = sales , hue = "Order Priority" , col ="Customer Segment" , col_wrap = 2)
```

```python
sns.relplot(x= "Sales" , y= "Profit" , data = sales , hue = "Ship Mode" , col ="Customer Segment" , col_wrap = 2)
```

```python
sns.relplot(x= "Sales" , y= "Profit" , data = sales , hue = "Ship Mode" , col ="Customer Segment" , row = "Product Category")
```
