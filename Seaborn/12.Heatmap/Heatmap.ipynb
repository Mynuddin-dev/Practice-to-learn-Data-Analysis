```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

sales = pd.read_excel("Sample - Superstore Sales.xls")
sales.head(2)
```

```python
# All code from seaborn heatmap documentation 

uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data)
```

```python
# Change the limits of the colormap:

ax = sns.heatmap(uniform_data, vmin=0, vmax=1)
```

```python
# Plot a heatmap for data centered on 0 with a diverging colormap:

normal_data = np.random.randn(10, 12)
ax = sns.heatmap(normal_data, center=0)
```

```python
flights = sns.load_dataset("flights")
flights.head(2)
```

```python
# Plot a dataframe with meaningful row and column labels:

flights = flights.pivot(index = "month", columns = "year", values = "passengers")   # pivot(index=None, columns=None, values=None) 
flights
```

```python
Ssns.heatmap(flights)
```

```python
# Annotate each cell with the numeric value using integer formatting:
# annotbool or rectangular dataset, optional
#If True, write the data value in each cell. If an array-like with the same shape as data, then use this to annotate the heatmap instead of the data. Note that DataFrames will match on position, not index.
# you can just keep cursor in function and press shift+tab

sns.heatmap(flights, annot=True, fmt="d")   # fmt -> String formatting code to use when adding annotations.
```

```python
# Add lines between each cell:

sns.heatmap(flights, linewidths=.5)
```

```python
# Use a different colormap:

sns.heatmap(flights, cmap="YlGnBu")
```

```python
# Center the colormap at a specific value:

sns.heatmap(flights, center=flights.loc["January", 1955])
```

```python
sns.heatmap(flights, center=0)
```


<p>xticklabels, yticklabels“auto”, bool, list-like, or int, optional
If True, plot the column names of the dataframe. If False, don’t plot the column names. If list-like, plot these alternate labels as the xticklabels. If an integer, use the column names but plot only every n label. If “auto”, try to densely plot non-overlapping labels.</p>

```python
# Plot every other column label and don’t plot row labels:

data = np.random.randn(50, 20)
sns.heatmap(data, xticklabels=2, yticklabels=False)
```

```python
# Don’t draw a colorbar:

sns.heatmap(flights, cbar=False)
```

```python
# Use different axes for the colorbar:
# In matplotlib discus about subplots

grid_kws = {"height_ratios": (.9, .05), "hspace": .3}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)

ax = sns.heatmap(flights, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"})
```
