```python
# seaborn.pydata.org

import seaborn as sns;
sns.set()
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
# Group by another variable and show the groups with different colors:


ax = sns.scatterplot(x="total_bill", y="tip", hue="time",data=tips)
```

```python
# Show the grouping variable by varying both color and marker:

ax = sns.scatterplot(x="total_bill", y="tip", hue="time", style="time", data=tips)
```

```python
# Vary colors and markers to show two different grouping variables:

ax = sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
```

```python
tips.head()
```

```python
# Show a quantitative variable by varying the size of the points:

ax = sns.scatterplot(x="total_bill", y="tip", size="size" , data=tips)
```

```python
#ax = sns.scatterplot(x="total_bill", y="tip", hue="day", style="time",size="size" , data=tips)
tips["time"]
```

```python
# Also show the quantitative variable by also using continuous colors:

ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", data=tips)
```

```python
# Use a different continuous color map:

cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)

ax = sns.scatterplot(x="total_bill" , y="tip", hue="size" , size="size", palette=cmap , data=tips)
```

```python
#Change the minimum and maximum point size and show all sizes in legend:

cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)

ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", sizes=(20, 200), palette=cmap, data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)

ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", sizes=(20, 200), palette=cmap,legend="full", data=tips)
```

```python
tips.columns
```

```python
# Vary the size with a categorical variable, and use a different palette:

ax = sns.scatterplot(x="total_bill", y="tip", hue="day", size="smoker", data=tips)  # here size is categorical variable
```

```python
# Use a specific set of markers:

markers = {"Lunch": "s", "Dinner": "X"}
ax = sns.scatterplot(x="total_bill", y="tip" , style="time" , markers=markers , data=tips)
```

```python
# Use relplot() to combine scatterplot() and FacetGrid: This allows grouping within additional categorical variables.
# Using relplot() is safer than using FacetGrid directly, 
# as it ensures synchronization of the semantic mappings across facets.

ax = sns.relplot(x="total_bill", y="tip", hue="day", size="smoker", data=tips)  # here size is categorical variable
```

```python

g = sns.relplot(x="total_bill", y="tip",col="time", hue="day", style="day",kind="scatter", data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)

ax = sns.relplot(x="total_bill", y="tip", hue="size", size="size", sizes=(20, 200), palette=cmap,legend="full", data=tips)
```
