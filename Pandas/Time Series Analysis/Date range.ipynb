```python
import pandas as pd
df = pd.read_csv("AAPL_no_dates.csv")
df
```

```python
rang = pd.date_range(start="6/1/2020",end="6/30/2020",freq='B') # B - > Business days which will exclude the weekend
rang
```

```python
df.set_index(rang, inplace=True)
df.head(10)
```

```python
%matplotlib inline
df.Close.plot()
```

```python
%matplotlib inline
df.Volume.plot()
```
