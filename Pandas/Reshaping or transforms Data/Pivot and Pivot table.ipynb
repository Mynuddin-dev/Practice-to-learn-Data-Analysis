
<h4>Pivot allows you to transforms or reshape datset</h4>

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_for_pivot.csv")
df
```

```python
df.pivot(index="Day"  ,columns = "city")
```

```python
df.pivot(index='city',columns='Day')
```

```python
df
```

```python
df.pivot(index='city',columns='Day',values="humidity")
```

```python
df.pivot(index='city',columns='Day',values="temperature")
```

```python
df.pivot(index='humidity',columns='city')  # that is also possible
```

<h2> Pivot table </h2>
<p> Pivot table used to summerized and aggregate data inside datframe </p>

```python
df = pd.read_csv("weather_for_pivot_table.csv")
df
```

```python
df.pivot_table(index = "city" , columns = "Day")
```

```python
df.pivot_table(index = "city" , columns = "Day" , aggfunc = "sum")
```

```python
df.pivot_table(index = "city" , columns = "Day" , aggfunc = "count" )
```

```python
df.pivot_table(index = "city" , columns = "Day" , aggfunc = "diff")
```

```python
df.pivot_table(index = "city" , columns = "Day" , aggfunc = "mean")
```

```python
df.pivot_table(index = "city" , columns = "Day" , margins = True)
```

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],columns=['C'], aggfunc=np.sum)
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],aggfunc={'D': np.mean,'E': [min, max, np.mean]})
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],aggfunc={'D':  [min, max, np.mean],'E': [min, max, np.mean]})
table
```
