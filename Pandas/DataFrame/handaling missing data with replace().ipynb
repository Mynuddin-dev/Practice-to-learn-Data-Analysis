```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_missin_replace().csv")
df
```

#Replacing single value

```python
new_df = df.replace(-99999, value=np.NaN)
new_df
```

#replacing list of value with single value

```python
new_df = df.replace(to_replace=[-99999,-9999], value=0)
new_df
```

```python
new_df = df.replace(to_replace=[-99999,-9999], value=np.NaN)
new_df
```


#replacing values based on specific column

```python
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -9999,
        'event': '0'
        }, np.NaN)
new_df
```


#Replacing by using mapping



```python
df
```

```python
new_df = df.replace({
        -99999: np.nan,
        -9999: np.nan,
        '0': 'Sunny',
    })
new_df
```

# Regex

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace('[A-Za-z]','', regex=True) 
new_df
```

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({
    'temperature': '[A-Za-z]', 
    'windspeed': '[a-z]'},'',
     regex=True) 
new_df
```
