```python
from collections import Counter
import csv
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")

data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))
    #print(language_counter)
                                                                            # i just wants to looks better and add space percentages
languages = []
popularity = []

for item in language_counter.most_common():
    languages.append(item[0])
    popularity.append(item[1])
    

languages.reverse()
popularity.reverse()

plt.barh(languages , popularity , color ="b" , align='center')

plt.title("Most Popular Languages")
plt.ylabel("Programing Languages")
plt.xlabel("No of Pepole Who use")
```
