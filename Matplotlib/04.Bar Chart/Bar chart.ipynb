```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

 <h3>Simple bar chart showing revenues of major US tech companies</h3>

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos , revenue)
```

```python
plt.xticks(xpos ,  company)
plt.bar(xpos , revenue)
```

```python
plt.xticks(xpos ,  company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.bar(xpos , revenue)
```

```python
plt.xticks(xpos ,  company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.bar(xpos , revenue , label="Revenue")
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos,revenue)
plt.bar(xpos,profit)

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
```

```python
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit, label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue, label="Revenue") # minus .2
plt.bar(xpos+0.2,profit,label="Profit")  #plus.2

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.barh(xpos,revenue, label="Revenue")      # h means horizental

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

```python
plt.barh(xpos-0.2,revenue, label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```
