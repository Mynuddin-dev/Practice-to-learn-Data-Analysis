```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days , max_t)
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

<h3>Axes labels and chart title</h3>

```python
plt.xlabel("days")
plt.ylabel("Temparature")
plt.title("Weather")
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="Days/Max")
plt.plot(days, min_t, label="Days/Min")
plt.plot(days, avg_t, label="Days/Average")
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="Days/Max")
plt.plot(days, min_t, label="Days/Min")
plt.plot(days, avg_t, label="Days/Average")

plt.legend()
```
