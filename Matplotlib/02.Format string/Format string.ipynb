```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
# syntex fmt = '[color][marker][line]'
plt.plot(x, y, 'go--', linewidth=2, markersize=8)   # green colour ,circle marker, dashed linestyle
```

```python
# same results
plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=8)
```

```python
plt.plot(x, y, color='#008840', marker='o', linestyle='dashed', linewidth=2, markersize=8)  # you can use hexadecimal value
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b+--')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'bD--' , linewidth=2, markersize=8)
# check formate string doc https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'--r*' , linewidth=2, markersize=8) 

# so order doesnot matter
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'--r^' , linewidth=2, markersize=8 ,alpha =1)   # alpha property set the transperancy
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'--r^' , linewidth=2, markersize=8 ,alpha =0.5) 
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'--r^' , linewidth=2, markersize=8 ,alpha =0) 
```
