```python
from matplotlib import pyplot as plt

plt.style.use("ggplot")

explode = [0, 0, 0, 0.1, 0]

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
colors = ["red","magenta","green" ,"yellow" , "gray"]



plt.pie(slices , labels = labels , colors = colors ,explode  = explode ,
        wedgeprops={'edgecolor': 'black'} ,shadow = True , startangle = 90 ,
        autopct='%1.1f%%' , textprops = dict(color ="blue")) 


plt.legend(loc = "center" ,shadow =True ,fontsize = "small")

plt.title("My Awesome Pie Chart")

```

```python

from matplotlib import pyplot as plt

plt.style.use("ggplot")

explode = [0.2, 0, 0.2, 0.1, 0]

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
colors = ["red","magenta","green" ,"yellow" , "gray"]

plt.title("My Awesome Pie Chart")

wp = { 'linewidth' : 1, 'edgecolor' : "green" } 


plt.pie(slices , labels = labels , colors = colors ,explode  = explode ,
        wedgeprops=wp ,shadow = True , startangle = 90 ,
        autopct='%1.1f%%' , textprops = dict(color ="blue")) 

# Adding legend 

plt.legend(loc = "upper center" ,shadow =True ,fontsize = "small")
```
