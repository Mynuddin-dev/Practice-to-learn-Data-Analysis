```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y , color = "Green")
```

```python
plt.plot(x,y , color = "Red")
#https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html?highlight=plot#module-matplotlib.pyplot
#https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
```

```python
plt.plot(x,y , "bo") # plot x and y using blue circle markers
```

```python
 plt.plot(x,y , "bo",linestyle='dashed')
```

```python
 plt.plot(x,y , "bo",linestyle='dashed',linewidth=2)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=3, linestyle='dotted')
```
